## 2. Print all the teams which have started playing between 1930-1980. (5 points)

```python
#Get first 4 char of Debut column and assign it to "StartedYear"
sport_df["StartedYear"] = sport_df.Debut.str.slice(0, 4)

#Convert "StartedYear COlumn to Int64 Type"
sport_df["StartedYear"] = sport_df["StartedYear"].astype('int64')

#Print the teams which have started playing between 1930 to 1980
sport_df[((sport_df["StartedYear"] >= 1930) & (sport_df["StartedYear"] <= 1980))]\
                .sort_values('StartedYear', ascending=False)[["Team","Debut","StartedYear"]]
```

## 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

```python
def Goal_diff_count(df):
    df['Goal Difference']=df['GoalsFor']-df['GoalsAgainst']
    df_teamGoalDiff=df[['Team','Goal Difference']].sort_values('Goal Difference',ascending=False)
    maxGoalDiff=df_teamGoalDiff[df_teamGoalDiff['Goal Difference'] == df['Goal Difference'].max()]
    minGoalDiff=df_teamGoalDiff[df_teamGoalDiff['Goal Difference'] == df['Goal Difference'].min()]
    return df_teamGoalDiff, maxGoalDiff, minGoalDiff #Return tuple value

teamsGoalDiff,maxGoalDiff,minGoalDiff=Goal_diff_count(sport_df)



print("Each Team Goal Difference : \n  ",teamsGoalDiff)

print("Maximum Goal Difference : \n  ",maxGoalDiff)

print("Minium Goal Difference : \n  ",minGoalDiff)

```
